```python
import re
import requests

# Function to extract the IP address from email headers
def extract_ip_from_header(email_header):
    # Regular expression to find an IP address in the 'Received' field
    ip_pattern = re.compile(r"Received:.*\[(?P<ip>\d+\.\d+\.\d+\.\d+)\]")
    match = ip_pattern.search(email_header)

    if match:
        return match.group("ip")
    else:
        print("IP Address not found in the email header.")
        return None

# Function to get location information based on IP address
def get_location_by_ip(ip):
    # Use the ipinfo.io API to get location info
    api_url = f"https://ipinfo.io/{ip}/json"
    response = requests.get(api_url)

    if response.status_code == 200:
        location_data = response.json()
        location_info = {
            'IP': ip,
            'Location': location_data.get('city', 'Unknown'),
            'Region': location_data.get('region', 'Unknown'),
            'Country': location_data.get('country', 'Unknown'),
            'Organization': location_data.get('org', 'Unknown'),
        }
        return location_info
    else:
        return {"Error": "Unable to fetch location data"}

# Sample email header (you can extract this from the email you want to analyze)
email_header = """
Delivered-To: adarshkumarjiitacin@gmail.com
Received: by 2002:a05:651c:388:b0:300:de9d:a3a2 with SMTP id e8csp2837350ljp;
        Wed, 15 Jan 2025 23:56:22 -0800 (PST)
X-Google-Smtp-Source: AGHT+IFB8b+p8ne0QTVPbWib1TnlyAlAeAmH7l9u9AX0jexE91r1a+/s9od7kyZhjOVf8tJmC1dy
X-Received: by 2002:a05:6000:1845:b0:38a:906e:16d0 with SMTP id ffacd0b85a97d-38bec542848mr1382733f8f.21.1737014182227;
        Wed, 15 Jan 2025 23:56:22 -0800 (PST)
ARC-Seal: i=1; a=rsa-sha256; t=1737014182; cv=none;
        d=google.com; s=arc-20240605;
        b=FsYAnEQ7daZorTvQD2q0bS0+u1yO0lD/+eeRMukYRNZ/zChEfE6wSbaodioCvH/B8l
         8w4FK+zR0dg1wNM/jqtRw8hwnId7tGGBD3XASkIOy9KY6b3RXGcEA4ZMt8BHWaIdXUeQ
         I3EpPiJQZZbbh180hOv2WuwqjIa2mUrCZQjGwDmywemdhvnANTLS+lmdyszdQgOPgPbt
         tvIY1xrL+8iY0G10WUKW1AA4iFYeKufA+ZAZ9X1+qSEj5BKpIwGjrsSQE0a2oneeoxmq
         orP4BVUJRb07OQmdeDF+s39np8ZV1wZsirI6kYbAza9UxkzbqTjGj38sqAYmet/1A6av
         t3xg==
ARC-Message-Signature: i=1; a=rsa-sha256; c=relaxed/relaxed; d=google.com; s=arc-20240605;
        h=mime-version:list-unsubscribe:subject:date:to:reply-to
         :origin-messageid:message-id:from:feedback-id:list-unsubscribe-post
         :dkim-signature;
        bh=2/DWkIOPxjOUW8zqSoLmflCd1Kostj4JojhIVOtJB1Y=;
        fh=TVrscxSecmPes9v4zCmHippjaqjLPdBgLn/Ybxpq654=;
        b=FzpuZeN8YXRPBWOZQZym4Ub+12yhS/v4dMd1V7DTRYGrt6tqEet8Zw2wtOqJ1wGqgU
         /JA75Ut0cPZRX2TenH9gomtOqD5ECYUaLyFLBsoW5J0O81fqrf1WayUmwqAhyUWNMhpZ
         VZbvfMZ16rq4oaCmHAgFlrJk4evAr9eI7x+YSoJ7fCtvZ3MgN4htggSMUC4FqAxkA0q/
         P7EiCCZDaMdFR0MMD1G75X/fQlfKrODgCMittdAJCDKF6KLiz7KxaAqd8mOLJUdirh95
         8ZS677ypv5pPWt0lsjNkt/NgQLUkC6DCHkwUwbTI6BbUXWpegMJ+UseK/8TVbJQs3nUj
         1dpg==;
        dara=google.com
ARC-Authentication-Results: i=1; mx.google.com;
       dkim=pass header.i=@dynamite.ai header.s=mail header.b=zPgjUFab;
       spf=pass (google.com: domain of bounces-293470714-1032254362@gf.d.sender-sib.com designates 77.32.148.6 as permitted sender) smtp.mailfrom=bounces-293470714-1032254362@gf.d.sender-sib.com;
       dmarc=pass (p=NONE sp=NONE dis=NONE) header.from=dynamite.ai
Return-Path: <bounces-293470714-1032254362@gf.d.sender-sib.com>
Received: from gf.d.sender-sib.com (gf.d.sender-sib.com. [77.32.148.6])
        by mx.google.com with ESMTPS id ffacd0b85a97d-38a8e5decc9si11448388f8f.865.2025.01.15.23.56.22
        for <adarshkumarjiitacin@gmail.com>
        (version=TLS1_3 cipher=TLS_AES_256_GCM_SHA384 bits=256/256);
        Wed, 15 Jan 2025 23:56:22 -0800 (PST)
Received-SPF: pass (google.com: domain of bounces-293470714-1032254362@gf.d.sender-sib.com designates 77.32.148.6 as permitted sender) client-ip=77.32.148.6;
Authentication-Results: mx.google.com;
       dkim=pass header.i=@dynamite.ai header.s=mail header.b=zPgjUFab;
       spf=pass (google.com: domain of bounces-293470714-1032254362@gf.d.sender-sib.com designates 77.32.148.6 as permitted sender) smtp.mailfrom=bounces-293470714-1032254362@gf.d.sender-sib.com;
       dmarc=pass (p=NONE sp=NONE dis=NONE) header.from=dynamite.ai
DKIM-Signature: v=1; a=rsa-sha256; c=relaxed/relaxed; d=dynamite.ai; q=dns/txt; s=mail; bh=2/DWkIOPxjOUW8zqSoLmflCd1Kostj4JojhIVOtJB1Y=; h=from:reply-to:subject:date:to:mime-version:content-type:list-unsubscribe:x-csa-complaints:list-unsubscribe-post:message-id:x-sib-id:feedback-id;
        b=zPgjUFabjXbHmbjuRdaWxWs/qm9ADeHV3GP6fbcRmWBoTcg0AiFl+KrW8DDc1LNq7DUsXjytLQH9
        bE3/8QUb5G3ZcXxEpXpTgTxCjUS6b82bTTLBp7vIUxIstvj4C7shqj1Dvx0A4FZmCp5xvUB1GzGQ
        r//RBg/7ULoHOgK/wkI=
X-Mailin-EID: MjkzNDcwNzE0fmFkYXJzaGt1bWFyamlpdGFjaW5AZ21haWwuY29tfjwyMDI1MDExNjA3NTYuMjI5MjUyNDM1MTNAc210cC1yZWxheS5tYWlsaW4uZnI%2BfmdmLmQuc2VuZGVyLXNpYi5jb20%3D
List-Unsubscribe-Post: List-Unsubscribe=One-Click
Feedback-ID: 77.32.148.6:8234870_-1:8234870:Sendinblue
From: DynamiteLab <noreply@dynamite.ai>
X-sib-id: Ks8Glu-QTStvDrMwiUReP6BG2s8idl5RtR9pX_eoxiugPHQ2PO6cExaT58Ijd_MJLfh2sX-Saxeqvd8ZLXeJgsUMVVW8p-9vaBRy4iclw1bx_9bb0KTSx0pFT2Y_6oIFLFVwaC75BCN4kWn5RVZgt6ZjhNUhMAeq47jBcliOPgFc8hbLlxEdKIQUMg
Message-Id: <fc11032c-1851-4965-b18f-e9f894afb21b@smtp-relay.sendinblue.com>
Origin-messageId: <202501160756.22925243513@smtp-relay.mailin.fr>
Reply-To: <info@dynamite.ai>
X-CSA-Complaints: csa-complaints@eco.de
To: <adarshkumarjiitacin@gmail.com>
Date: Thu, 16 Jan 2025 07:56:21 +0000
Subject: Welcome to DynamiteLab
List-Unsubscribe: <https://icdeiha.r.bh.d.sendibt3.com/tr/un/li/F32StHOgItiITzUAXHHWggT11C9IweDE_8GEvwuSaiWAUuVT-W7KJALVOdFRAD4JWijl0kbMPVIpURhWymljYdf2sN2EE0HD1V4xXYQwB1a0A9QqunZZ9nz7IysQE7vLgjDrAspfaGYq06KWI62NEZSaJLRmPyn1a3SwD-dG8_S9X_MlPMRyb6xbxVmu_513SxcmAh61kZlhikkL1VaImsNMr7IM0qKz6mmxPBuZ1mN3mehjeEJGzqS0bw>
Content-Type: multipart/alternative; boundary="--_SiB-39bb7901476226a7-Part_1"
X-Mailin-Tag: brevo_signup
Mime-Version: 1.0

----_SiB-39bb7901476226a7-Part_1
Content-Transfer-Encoding: quoted-printable
Content-Type: text/plain; charset=utf-8

Greetings!
=C2=A0
Thank you for signing up with [DynamiteLab](https://lab.dynamite.ai/). Sinc=
e the inception of PacketTotal in 2017, our team has served cybersecurity p=
rofessionals and has built a research repository of over 120,000 network tr=
affic samples. In 2023, DynamiteLab Community became the successor to Packe=
tTotal, serving as=C2=A0a free,=C2=A0public workspace that openly shares PC=
AP files with all of its members.=C2=A0

=C2=A0

We would like to remind you NOT to upload sensitive PCAP data to the Commun=
ity workspace. Instead, you can set up a private workspace with restricted =
access for a small subscription fee. [Let us know](https://lab.dynamite.ai/=
contact) if you have any feedback or if we can help with your enterprise ne=
eds.=C2=A0

=C2=A0

Best,
=C2=A0- Oleg
=C2=A0
Oleg Sinitsin
CEO & Founder, [Dynamite Analytics](https://dynamite.ai/)
Packet Capture Intelligence with [DynamiteLab](https://lab.dynamite.ai/)

You've received this email because you signed up with DynamiteLab

Dynamite Analytics =C2=B7 7742 Spalding Drive #359 =C2=B7 Norcross, GA 3009=
2 =C2=B7 USA

[Unsubscribe](https://icdeiha.r.bh.d.sendibt3.com/tr/un/pcNKyPMRYjZvV_lU_76=
s5zcpdEFdVGr_lYwBCb1hj1iciHgd8t5bL4LFkhn4Yf6c6cpFFeUZY7NCB7suMlFGnZWmhviNOy=
Bpxldeqdrn0VVJZpt3ks353BnhcITY9E-yEAMVk5E_i2j12s1EYTqC9l6KxhxoZqnbqBGV86vo_=
N4kfp5zO3D6f7nEw86BhECN3NQuhKSX2W_dK-eD6jGpYUf8c3cJcRLn4TLViDyFipFrBQqzlK-i=
OOPOg97tCQZ_XK_xjkJcCV68snMLluAJSu-kFeqaVgYDd9V2)


----_SiB-39bb7901476226a7-Part_1
Content-Transfer-Encoding: quoted-printable
Content-Type: text/html; charset=utf-8

<!DOCTYPE html PUBLIC "-//W3C//DTD XHTML 1.0 Strict//EN" "http://www.w3.org=
/TR/xhtml1/DTD/xhtml1-strict.dtd"><html xmlns=3D"http://www.w3.org/1999/xht=
ml" xmlns:v=3D"urn:schemas-microsoft-com:vml" xmlns:o=3D"urn:schemas-micros=
oft-com:office:office"><head><meta http-equiv=3D"Content-Type" content=3D"t=
ext/html; charset=3Dutf-8"/><meta http-equiv=3D"X-UA-Compatible" content=3D=
"IE=3Dedge"/><meta name=3D"format-detection" content=3D"telephone=3Dno"/><m=
eta name=3D"viewport" content=3D"width=3Ddevice-width, initial-scale=3D1.0"=
/><title>Welcome to DynamiteLab</title><style type=3D"text/css" emogrify=3D=
"no">#outlook a { padding:0; } .ExternalClass { width:100%; } .ExternalClas=
s, .ExternalClass p, .ExternalClass span, .ExternalClass font, .ExternalCla=
ss td, .ExternalClass div { line-height: 100%; } table td { border-collapse=
: collapse; mso-line-height-rule: exactly; } .editable.image { font-size: 0=
 !important; line-height: 0 !important; } .nl2go_preheader { display: none =
!important; mso-hide:all !important; mso-line-height-rule: exactly; visibil=
ity: hidden !important; line-height: 0px !important; font-size: 0px !import=
ant; } body { width:100% !important; -webkit-text-size-adjust:100%; -ms-tex=
t-size-adjust:100%; margin:0; padding:0; } img { outline:none; text-decorat=
ion:none; -ms-interpolation-mode: bicubic; } a img { border:none; } table {=
 border-collapse:collapse; mso-table-lspace:0pt; mso-table-rspace:0pt; } th=
 { font-weight: normal; text-align: left; } *[class=3D"gmail-fix"] { displa=
y: none !important; } </style><style type=3D"text/css" emogrify=3D"no"> @me=
dia (max-width: 600px) { .gmx-killpill { content: ' \03D1';} } </style><sty=
le type=3D"text/css" emogrify=3D"no">@media (max-width: 600px) { .gmx-killp=
ill { content: ' \03D1';} .r0-o { border-style: solid !important; margin: 0=
 auto 0 auto !important; width: 100% !important } .r1-i { background-color:=
 #ffffff !important } .r2-c { box-sizing: border-box !important; text-align=
: center !important; valign: top !important; width: 320px !important } .r3-=
o { border-style: solid !important; margin: 0 auto 0 auto !important; width=
: 320px !important } .r4-i { padding-bottom: 10px !important; padding-left:=
 15px !important; padding-right: 15px !important; padding-top: 10px !import=
ant } .r5-c { box-sizing: border-box !important; display: block !important;=
 valign: top !important; width: 100% !important } .r6-o { border-style: sol=
id !important; width: 100% !important } .r7-i { padding-left: 0px !importan=
t; padding-right: 0px !important } .r8-c { box-sizing: border-box !importan=
t; text-align: center !important; valign: top !important; width: 200px !imp=
ortant } .r9-o { border-style: solid !important; margin: 0 auto 0 auto !imp=
ortant; width: 200px !important } .r10-i { padding-bottom: 10px !important;=
 padding-top: 20px !important } .r11-c { box-sizing: border-box !important;=
 text-align: center !important; valign: top !important; width: 100% !import=
ant } .r12-c { box-sizing: border-box !important; padding-bottom: 10px !imp=
ortant; padding-top: 10px !important; text-align: center !important; width:=
 100% !important } .r13-c { box-sizing: border-box !important; padding-bott=
om: 15px !important; padding-top: 15px !important; text-align: left !import=
ant; valign: top !important; width: 100% !important } .r14-i { padding-bott=
om: 20px !important; padding-left: 15px !important; padding-right: 15px !im=
portant; padding-top: 20px !important } .r15-c { box-sizing: border-box !im=
portant; text-align: left !important; valign: top !important; width: 100% !=
important } .r16-o { border-style: solid !important; margin: 0 auto 0 0 !im=
portant; width: 100% !important } .r17-i { color: #3b3f44 !important; paddi=
ng-bottom: 0px !important; padding-top: 0px !important; text-align: center =
!important } .r18-i { color: #3b3f44 !important; padding-bottom: 5px !impor=
tant; padding-top: 5px !important; text-align: center !important } .r19-c {=
 box-sizing: border-box !important; text-align: center !important; width: 1=
00% !important } .r20-i { padding-bottom: 15px !important; padding-left: 0p=
x !important; padding-right: 0px !important; padding-top: 0px !important } =
.r21-c { box-sizing: border-box !important; text-align: center !important; =
valign: top !important; width: 129px !important } .r22-o { border-style: so=
lid !important; margin: 0 auto 0 auto !important; width: 129px !important }=
 body { -webkit-text-size-adjust: none } .nl2go-responsive-hide { display: =
none } .nl2go-body-table { min-width: unset !important } .mobshow { height:=
 auto !important; overflow: visible !important; max-height: unset !importan=
t; visibility: visible !important } .resp-table { display: inline-table !im=
portant } .magic-resp { display: table-cell !important } } </style><style t=
ype=3D"text/css">p, h1, h2, h3, h4, ol, ul, li { margin: 0; } a, a:link { c=
olor: #185bdc; text-decoration: underline } .nl2go-default-textstyle { colo=
r: #3b3f44; font-family: Lato, arial; font-size: 16px; line-height: 1.5; wo=
rd-break: break-word } .default-button { color: #ffffff; font-family: Lato,=
 arial; font-size: 16px; font-style: normal; font-weight: normal; line-heig=
ht: 1.15; text-decoration: none; word-break: break-word } .default-heading1=
 { color: #1F2D3D; font-family: Lato, arial; font-size: 36px; word-break: b=
reak-word } .default-heading2 { color: #1F2D3D; font-family: Lato, arial; f=
ont-size: 32px; word-break: break-word } .default-heading3 { color: #1F2D3D=
; font-family: Lato, arial; font-size: 24px; word-break: break-word } .defa=
ult-heading4 { color: #1F2D3D; font-family: Lato, arial; font-size: 18px; w=
ord-break: break-word } a[x-apple-data-detectors] { color: inherit !importa=
nt; text-decoration: inherit !important; font-size: inherit !important; fon=
t-family: inherit !important; font-weight: inherit !important; line-height:=
 inherit !important; } .no-show-for-you { border: none; display: none; floa=
t: none; font-size: 0; height: 0; line-height: 0; max-height: 0; mso-hide: =
all; overflow: hidden; table-layout: fixed; visibility: hidden; width: 0; }=
 </style><!--[if mso]><xml> <o:OfficeDocumentSettings> <o:AllowPNG/> <o:Pix=
elsPerInch>96</o:PixelsPerInch> </o:OfficeDocumentSettings> </xml><![endif]=
--><style type=3D"text/css">a:link{color: #185bdc; text-decoration: underli=
ne;}</style></head><body bgcolor=3D"#ffffff" text=3D"#3b3f44" link=3D"#185b=
dc" yahoo=3D"fix" style=3D"background-color: #ffffff;"><img width=3D"1" hei=
ght=3D"1" src=3D"https://icdeiha.r.bh.d.sendibt3.com/tr/op/6PEPVQseg89_t3bS=
Xv09B0pYar4kTVKV-hVxHeq8zE94277UZFSfEahXM5vrnC5DOiIaPMdHd8ZKArOpl1WCUskO93c=
9j4PAH6idWpWHZpIsXSX81dnxa3nlhl4CnaYw4zXuJDUgRFfpeUTN9IylXUi2slpkbqhbNPKbt0=
v05J-SnpuXwX7tS44pnqN9YLsuugAAYvzHP5ZaFsvvxgLVuxFaqa3PqFkYoXplflJGUaFY" sty=
le=3D"display:none"/><table cellpadding=3D"0" border=3D"0" cellspacing=3D"0=
" style=3D" mso-hide: all; display: none;"><tbody><tr><td>PCAP Analysis Rei=
nvented -</td></tr></tbody></table> <table cellspacing=3D"0" cellpadding=3D=
"0" border=3D"0" role=3D"presentation" class=3D"nl2go-body-table" width=3D"=
100%" style=3D"background-color: #ffffff; width: 100%;"><tbody><tr><td> <ta=
ble cellspacing=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presentation" =
width=3D"100%" align=3D"center" class=3D"r0-o" style=3D"table-layout: fixed=
; width: 100%;"><tbody><tr><td valign=3D"top" class=3D"r1-i" style=3D"backg=
round-color: #ffffff;"> <table cellspacing=3D"0" cellpadding=3D"0" border=
=3D"0" role=3D"presentation" width=3D"600" align=3D"center" class=3D"r3-o" =
style=3D"table-layout: fixed; width: 600px;"><tbody><tr><td class=3D"r4-i" =
style=3D"padding-bottom: 10px; padding-top: 10px;"> <table width=3D"100%" c=
ellspacing=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presentation"><tbod=
y><tr><th width=3D"100%" valign=3D"top" class=3D"r5-c" style=3D"font-weight=
: normal;"> <table cellspacing=3D"0" cellpadding=3D"0" border=3D"0" role=3D=
"presentation" width=3D"100%" class=3D"r6-o" style=3D"table-layout: fixed; =
width: 100%;"><tbody><tr><td valign=3D"top" class=3D"r7-i" style=3D"padding=
-left: 15px; padding-right: 15px;"> <table width=3D"100%" cellspacing=3D"0"=
 cellpadding=3D"0" border=3D"0" role=3D"presentation"><tbody><tr><td class=
=3D"r8-c" align=3D"center"> <table cellspacing=3D"0" cellpadding=3D"0" bord=
er=3D"0" role=3D"presentation" width=3D"200" class=3D"r9-o" style=3D"table-=
layout: fixed; width: 200px;"><tbody><tr><td class=3D"r10-i" style=3D"font-=
size: 0px; line-height: 0px; padding-bottom: 10px; padding-top: 20px;"> <a =
href=3D"https://icdeiha.r.bh.d.sendibt3.com/tr/cl/jZC6fo3qH9Vaum89rqgxZcXNv=
61vx5ne01bdEqrY7kgBEsSt-dWKPYrudgIMJ7qPvzv1QIafap2IuorZOlPU8-wUqQ_MskpH0iBv=
ja1EjfQQqDaYAqA1F7qtORdT-Hc4hL48Obmer6OPNt8_jY82yos-VP3Mtd-vXs7JHYUGSZkB0vV=
xtHoR4mzJFBilwVDtcgJ16wun39itwJwaatCBrprOERbJQ8tZEcrZipa3S7BJQsBKNOVbIUO04a=
m84p8qQTjukxXC4JYRPhUbt99Y8YMJXRMVAvhVucYfevMoXlU4D2ZrvqTe4117Hk8dx2IdFuraX=
Vch85BuY_hLz3-lcjwfy7nvmsLuJvkf9a5zJg9mO0hE0EVNVhumuYkGxEQQohl2jf-gHg2z-6KF=
o9GbFi3hDW3pqv_G" target=3D"_blank" style=3D"color: #185bdc; text-decoratio=
n: underline;"> <img src=3D"https://icdeiha.r.bh.d.sendibt3.com/im/8234870/=
7e093d5ad29e14c3ab98e4addb2a157c92e7535a709cf3af83b557382cd38d26.png?e=3DHY=
HFVu90kfom3LTBuwmvdhfdVWSQMYdfotu1h1UGjrAmmqskUe3dy1hn12BFrzU9_bwmSBEnh3ueV=
PnsaDKgGI1efxDZp3Jyy9OVlh2QUa5XwyEU-YGAbz-yGDEChNHDFaWjk_Ol6xWk556xO04J5Rci=
phf_4AB_KP0ap6ROSWj-k3_0lLWQkUTsvmelXVGSZjynbaMWZ68RmVNd5bK7fmRDjaAvQNaUK3m=
tt_TIoAxbotx-Q2aabA2Jtoi_U_2T865rG5jw2uSNyCSSIHjgmDLbU04lw_HNKw_d7SS3Shuw2q=
UJgmZbT_H-aLm2xAMED-9ne1W0wF6IYx8" width=3D"200" border=3D"0" style=3D"disp=
lay: block; width: 100%;"/></a> </td> </tr></tbody></table></td> </tr></tbo=
dy></table></td> </tr></tbody></table></th> </tr></tbody></table></td> </tr=
></tbody></table></td> </tr></tbody></table><table cellspacing=3D"0" cellpa=
dding=3D"0" border=3D"0" role=3D"presentation" width=3D"600" align=3D"cente=
r" class=3D"r3-o" style=3D"table-layout: fixed; width: 600px;"><tbody><tr><=
td valign=3D"top" class=3D"r1-i" style=3D"background-color: #ffffff;"> <tab=
le cellspacing=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presentation" w=
idth=3D"100%" align=3D"center" class=3D"r0-o" style=3D"table-layout: fixed;=
 width: 100%;"><tbody><tr><td class=3D"r4-i" style=3D"padding-bottom: 10px;=
 padding-top: 10px;"> <table width=3D"100%" cellspacing=3D"0" cellpadding=
=3D"0" border=3D"0" role=3D"presentation"><tbody><tr><th width=3D"100%" val=
ign=3D"top" class=3D"r5-c" style=3D"font-weight: normal;"> <table cellspaci=
ng=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presentation" width=3D"100%=
" class=3D"r6-o" style=3D"table-layout: fixed; width: 100%;"><tbody><tr><td=
 valign=3D"top" class=3D"r7-i" style=3D"padding-left: 15px; padding-right: =
15px;"> <table width=3D"100%" cellspacing=3D"0" cellpadding=3D"0" border=3D=
"0" role=3D"presentation"><tbody><tr><td class=3D"r12-c" align=3D"center" s=
tyle=3D"padding-bottom: 10px; padding-top: 10px;"> <table width=3D"100%" ce=
llspacing=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presentation" height=
=3D"1" style=3D"border-top-style: solid; background-clip: border-box; borde=
r-top-color: #9a9a9a; border-top-width: 1px; font-size: 1px; line-height: 1=
px;"><tbody><tr><td height=3D"0" style=3D"font-size: 0px; line-height: 0px;=
">=C2=AD</td> </tr></tbody></table></td> </tr></tbody></table></td> </tr></=
tbody></table></th> </tr></tbody></table></td> </tr></tbody></table><table =
cellspacing=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presentation" widt=
h=3D"100%" align=3D"center" class=3D"r0-o" style=3D"table-layout: fixed; wi=
dth: 100%;"><tbody><tr><td class=3D"r4-i" style=3D"padding-bottom: 10px; pa=
dding-top: 10px;"> <table width=3D"100%" cellspacing=3D"0" cellpadding=3D"0=
" border=3D"0" role=3D"presentation"><tbody><tr><th width=3D"100%" valign=
=3D"top" class=3D"r5-c" style=3D"font-weight: normal;"> <table cellspacing=
=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presentation" width=3D"100%" =
class=3D"r6-o" style=3D"table-layout: fixed; width: 100%;"><tbody><tr><td v=
align=3D"top" class=3D"r7-i" style=3D"padding-left: 15px; padding-right: 15=
px;"> <table width=3D"100%" cellspacing=3D"0" cellpadding=3D"0" border=3D"0=
" role=3D"presentation"><tbody><tr><td class=3D"r13-c nl2go-default-textsty=
le" align=3D"left" style=3D"color: #3b3f44; font-family: Lato,arial; font-s=
ize: 16px; line-height: 1.5; word-break: break-word; padding-bottom: 15px; =
padding-top: 15px; text-align: left; valign: top;"> <div><p style=3D"margin=
: 0; text-align: justify;">Greetings!<br/>=C2=A0<br/>Thank you for signing =
up with <a href=3D"https://icdeiha.r.bh.d.sendibt3.com/tr/cl/Z6WK_b8xMqJFXS=
GKr6MP4uPQiBr05YDbqJajaI4lLHHDX1sjdslaRg3i3wuoG76RN4sH_OV6XS9BHTqDzEgPHQexh=
ArRk0X1RngfsDn8KDpN8c3qVjHwqSwjR8PJTYMSnNg2ig0dkZj3E6EHDYZ-qZSTOgcav2TrwkXE=
Y0xbprP6KKSiAEJLxyoye6HFEfqojYrcmfu3NVx6tDSoEJix0eY5gFm2vuNEY_gnJoq41T8316z=
tGRwd4Qyz5Dun5eYjL3ILGnZakt9G4gMsGi91ZStVqS8eLsnxcIfHSKZ_jgCiP-QJsD7rq5MdFU=
razSunRgfLEB8koKEfgMEf24udL5W9W-SCzI6AnQfaWhvyGOQZbanEyJr3YuHLqY--XOBhkZ_ei=
Eu6TZKsj6xMRssZdlALuczmM-KtYoHgeQ" title=3D"" target=3D"_blank" style=3D"co=
lor: #185bdc; text-decoration: underline;">DynamiteLab</a>. Since the incep=
tion of PacketTotal in 2017, our team has served cybersecurity professional=
s and has built a research repository of over 120,000 network traffic sampl=
es. In 2023, DynamiteLab Community became the successor to PacketTotal, ser=
ving as=C2=A0a free,=C2=A0public workspace that openly shares PCAP files wi=
th all of its members.=C2=A0</p><p style=3D"margin: 0; text-align: justify;=
">=C2=A0</p><p style=3D"margin: 0; text-align: justify;">We would like to r=
emind you NOT to upload sensitive PCAP data to the Community workspace. Ins=
tead, you can set up a private workspace with restricted access for a small=
 subscription fee. <a href=3D"https://icdeiha.r.bh.d.sendibt3.com/tr/cl/3b9=
AeD-Xa-97ZiANZspAWSiQsCJ0RLBy9d6ZNLDCbl6OjrjWU0GYhwoDBJau-_3VsDSSNy0GuisQG2=
SCa24x5LgAyqPk_mL48GXirteRytXe9XhD-tayLO1w0E354xFORx10ceHm_Kt4AUhEy79B454m5=
ZunU-GcnTEL3ZnoJe08TtWIqj9NX-qt1ICtrEk5hzzVIHqKdqIjNTmCD-HUUUPN74YvezSvb4i2=
uYtk1AoVs9NhUdPlJdbpzqPBMKJyuJTMDsHRTmlipUBIaZnLuZgzwnO5dpd4VD1oRqzqsb--iR5=
CLAq2RyHmRmxo5NQOOK7GCk4MoDdg9aecjZn48-YAv0rlhY3Kqh1Rp2YEoPgqVD2BW-2x6xnRep=
IN-89Tj5hMVTcCg-IPIKzzGzstXbWxElr9z4DFruNeqsZQuuMbJzg" target=3D"_blank" st=
yle=3D"color: #185bdc; text-decoration: underline;">Let us know</a> if you =
have any feedback or if we can help with your enterprise needs.=C2=A0</p><p=
 style=3D"margin: 0;">=C2=A0</p><p style=3D"margin: 0;">Best,<br/>=C2=A0- O=
leg<br/>=C2=A0<br/>Oleg Sinitsin<br/>CEO &amp; Founder, <a href=3D"https://=
icdeiha.r.bh.d.sendibt3.com/tr/cl/VxuvM1O-a4vJtb5qQzgbPGT11D3W4kS3oCHtusd6W=
gPlXzP8vY0xV5iPqtdXniwByNcLojuo-5qgQdB-MQl4pzEyexlWqLkUDakofD2HR9Qm_Hq41VFi=
dhu7SN5yKbIlQgPtdqgFu3uh41D1NM67xYYkenOV0drrm5CAr2NA8pjpesQDhikpxyVSp_DXHrE=
BaLCjwWi4qqv20IYROlMgN9pOz6AFjZQ3Z7Rew8kBPlbXbhwmANnYAzEqvN5Mk6SWfBnlxqc-Uj=
ud4EsQUYM2GxlTyHXu5QKJMEOJnC6ra8nkZ3mw-rfvBSu7oeXMfwB2wWYwy7MupNuwI5S09ZFQv=
mxRZJ8jmqJCeSBCUfdPaYu8gP157rThduVlixU7yK-JhbGkE6WyK6faa6wfbyRuq83BOadaKb79=
" target=3D"_blank" style=3D"color: #185bdc; text-decoration: underline;">D=
ynamite Analytics</a><br/>Packet Capture Intelligence with <a href=3D"https=
://icdeiha.r.bh.d.sendibt3.com/tr/cl/DFEQOHWK4eNpAAHDzyOqV8Lajz1P0vgxj_lMD3=
CwjMIl6NI_rUJa6_fZ8ufyJtv_D4D-AB1guCgMx_6tK4FTpqvJD03FFIPPPRwp9LhDcQP6_N6hP=
iqdWYHGzjnCwdG6QyNUckchawA2zBElOOXrZVaMLXXAl-oyAM842x8GICNk150E9y__AEiO-Lj1=
ouoNFGnXuNqNvYCqBj-xYsCXVLKE5L3ttJPkghDhOvD_nvQA4DvWgKVcL17LSdgEi0aP1ygd53D=
i3Nnxd27Y3JAEW-3dmAWqHRWw76i4ORgOIgDa5wHodtHS8pG2HCY7v_XPrNX9t0nNFQPVb_eYBd=
cU2UiJDYwDr2tGIRYpmA7p9mPP8n1I3V-MS-u146EJuFeJcQhNFaFFZHBTS024lL4MsAzyc6e6k=
fSqb6Tv2w" target=3D"_blank" style=3D"color: #185bdc; text-decoration: unde=
rline;">DynamiteLab</a></p></div> </td> </tr></tbody></table></td> </tr></t=
body></table></th> </tr></tbody></table></td> </tr></tbody></table><table c=
ellspacing=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presentation" width=
=3D"100%" align=3D"center" class=3D"r0-o" style=3D"table-layout: fixed; wid=
th: 100%;"><tbody><tr><td class=3D"r14-i" style=3D"padding-bottom: 20px; pa=
dding-top: 20px;"> <table width=3D"100%" cellspacing=3D"0" cellpadding=3D"0=
" border=3D"0" role=3D"presentation"><tbody><tr><th width=3D"100%" valign=
=3D"top" class=3D"r5-c" style=3D"font-weight: normal;"> <table cellspacing=
=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presentation" width=3D"100%" =
class=3D"r6-o" style=3D"table-layout: fixed; width: 100%;"><tbody><tr><td v=
align=3D"top" class=3D"r7-i" style=3D"padding-left: 15px; padding-right: 15=
px;"> <table width=3D"100%" cellspacing=3D"0" cellpadding=3D"0" border=3D"0=
" role=3D"presentation"><tbody><tr><td class=3D"r12-c" align=3D"center" sty=
le=3D"padding-bottom: 10px; padding-top: 10px;"> <table width=3D"100%" cell=
spacing=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presentation" height=
=3D"1" style=3D"border-top-style: solid; background-clip: border-box; borde=
r-top-color: #9a9a9a; border-top-width: 1px; font-size: 1px; line-height: 1=
px;"><tbody><tr><td height=3D"0" style=3D"font-size: 0px; line-height: 0px;=
">=C2=AD</td> </tr></tbody></table></td> </tr></tbody></table></td> </tr></=
tbody></table></th> </tr></tbody></table></td> </tr></tbody></table></td> <=
/tr></tbody></table><table cellspacing=3D"0" cellpadding=3D"0" border=3D"0"=
 role=3D"presentation" width=3D"100%" align=3D"center" class=3D"r0-o" style=
=3D"table-layout: fixed; width: 100%;"><tbody><tr><td valign=3D"top" class=
=3D"r1-i" style=3D"background-color: #ffffff;"> <table cellspacing=3D"0" ce=
llpadding=3D"0" border=3D"0" role=3D"presentation" width=3D"600" align=3D"c=
enter" class=3D"r3-o" style=3D"table-layout: fixed; width: 600px;"><tbody><=
tr><td class=3D"r14-i" style=3D"padding-bottom: 20px; padding-top: 20px;"> =
<table width=3D"100%" cellspacing=3D"0" cellpadding=3D"0" border=3D"0" role=
=3D"presentation"><tbody><tr><th width=3D"100%" valign=3D"top" class=3D"r5-=
c" style=3D"font-weight: normal;"> <table cellspacing=3D"0" cellpadding=3D"=
0" border=3D"0" role=3D"presentation" width=3D"100%" class=3D"r6-o" style=
=3D"table-layout: fixed; width: 100%;"><tbody><tr><td valign=3D"top" class=
=3D"r7-i" style=3D"padding-left: 15px; padding-right: 15px;"> <table width=
=3D"100%" cellspacing=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presenta=
tion"><tbody><tr><td class=3D"r15-c" align=3D"left"> <table cellspacing=3D"=
0" cellpadding=3D"0" border=3D"0" role=3D"presentation" width=3D"100%" clas=
s=3D"r16-o" style=3D"table-layout: fixed; width: 100%;"><tbody><tr><td alig=
n=3D"center" valign=3D"top" class=3D"r17-i nl2go-default-textstyle" style=
=3D"font-family: Lato,arial; word-break: break-word; color: #3b3f44; font-s=
ize: 18px; line-height: 1.5; text-align: center;"> <div><p style=3D"margin:=
 0; font-size: 14px;"><span style=3D"font-size: 11px;">You&#39;ve received =
this email because you signed up with DynamiteLab</span></p></div> </td> </=
tr></tbody></table></td> </tr><tr><td class=3D"r15-c" align=3D"left"> <tabl=
e cellspacing=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presentation" wi=
dth=3D"100%" class=3D"r16-o" style=3D"table-layout: fixed; width: 100%;"><t=
body><tr><td align=3D"center" valign=3D"top" class=3D"r17-i nl2go-default-t=
extstyle" style=3D"font-family: Lato,arial; word-break: break-word; color: =
#3b3f44; font-size: 18px; line-height: 1.5; text-align: center;"> <div><p s=
tyle=3D"margin: 0; font-size: 14px;"><span style=3D"font-size: 11px;">Dynam=
ite Analytics =C2=B7 7742 Spalding Drive #359 =C2=B7 Norcross, GA 30092 =C2=
=B7 USA</span></p></div> </td> </tr></tbody></table></td> </tr><tr><td clas=
s=3D"r15-c" align=3D"left"> <table cellspacing=3D"0" cellpadding=3D"0" bord=
er=3D"0" role=3D"presentation" width=3D"100%" class=3D"r16-o" style=3D"tabl=
e-layout: fixed; width: 100%;"><tbody><tr><td align=3D"center" valign=3D"to=
p" class=3D"r18-i nl2go-default-textstyle" style=3D"font-family: Lato,arial=
; word-break: break-word; color: #3b3f44; font-size: 18px; line-height: 1.5=
; padding-bottom: 5px; padding-top: 5px; text-align: center;"> <div><p styl=
e=3D"margin: 0; font-size: 14px;"><a href=3D"https://icdeiha.r.bh.d.sendibt=
3.com/tr/un/aMXk7cWbLnHgxGjeSS_5uoni36ASWZKzzPH1RM36BnX_xsPNblG4VC43sOz9OX4=
4Vb8V95Nt7_gJzgPxzZU2Zpj81vgxbZ8eVTmY0ly-Uzz-bP0Zgd1A4wN6r04QLZiZlTwtGxdD_o=
3VfXE9xiE3hrBCYgtWLb3Lj0heP2ct-eZ_K_TefkLMDFQ2PvJLr1bm8P8Rx07TZ0-P3-TUtVrUe=
mLAtETPMaRip5SsjtRq-JtyInwm4mrqBn3DIv0Fbrytx7XB1apspyX8Iu19Om0rDwee06jqxtm1=
SpQA" target=3D"_blank" style=3D"color: #185bdc; text-decoration: underline=
;"><span style=3D"font-size: 11px;">Unsubscribe</span></a></p></div> </td> =
</tr></tbody></table></td> </tr><tr><td class=3D"r19-c" align=3D"center"> <=
table cellspacing=3D"0" cellpadding=3D"0" border=3D"0" role=3D"presentation=
" width=3D"100%" align=3D"center" class=3D"r0-o" style=3D"table-layout: fix=
ed; width: 100%;"><tbody><tr><td valign=3D"top" class=3D"r20-i" style=3D"pa=
dding-bottom: 15px;"> <table width=3D"100%" cellspacing=3D"0" cellpadding=
=3D"0" border=3D"0" role=3D"presentation"><tbody><tr><td class=3D"r21-c" al=
ign=3D"center"> <table cellspacing=3D"0" cellpadding=3D"0" border=3D"0" rol=
e=3D"presentation" width=3D"129" class=3D"r22-o" style=3D"table-layout: fix=
ed;"><tbody><tr><td height=3D"48" style=3D"font-size: 0px; line-height: 0px=
;"> <a href=3D"https://icdeiha.r.bh.d.sendibt3.com/tr/cl/icqD-7AEiREfp__stb=
2iMqCoP6-7qtQEVZEYk6qqLsoBnSpT_V3XSFeUiHzyrOLvWs5vSig4bogSNlk51MCeiuPAj0nRD=
C8vjkf0_xSVkg6Z2GHIkR57AsE94Ej1YFlIOh6Gv1G2DV1lZZlw9upR5mcOsc_gnhPEbDN0rnp6=
UYLavQ25IHpG5jXQ30-TIJtBODLqWcx1ZiEtJfNCm2GUN9DiOn63R8vZbodktxrR3UJkzbPRMBJ=
35GIznV4R3sfho2FeXyMhpR5Op_AbsPuRm4k5-P0hZg9kTjL55nDTluWOgLOleZcsgor_g1SNV8=
U_UNn-PcTAYPwd65IvQG5AQPBt5HNMNRxbYqiEklFrZmLvBRREMrVHxStehUy0E4cZzFUajAbmr=
oKYvsJZ2rh8-ougl9P25-s"><img src=3D"https://icdeiha.r.bh.d.sendibt3.com/im/=
8234870/ec9bdbff369bfa6eded87bb22dd8c4f320454721e5daf3e5b5ee5091a2ffc8f1.pn=
g?e=3De_CjknUO2ENkNoDPbXOJd1ZG6vtn7-QAhfPjhT9PQQWKDr3lqj6saSuG5SBUEW8ULAeBx=
kpWyisiF-2ZSUNSXJCLD3aq3TbIeAbCXEAF0eHldeLn4W8hhqMESNXqpgBa6MRi-LyEiVozBLQr=
S4SSPchjVkZxKJ4rleQ_93GBowagOX6PruJ8aimqWDgq6OnxZOxNOEkkeZz4KK8VtSa3whMCcT7=
w3nMeYOsIfT6AaZSYQf39U-MDNTMfSzY0bpLkqGb8oTt5ocTKaG7fJ1g" width=3D"129" hei=
ght=3D"48" border=3D"0" style=3D"display: block; width: 100%;"/></a></td> <=
/tr></tbody></table></td> </tr></tbody></table></td> </tr></tbody></table><=
/td> </tr></tbody></table></td> </tr></tbody></table></th> </tr></tbody></t=
able></td> </tr></tbody></table></td> </tr></tbody></table></td> </tr></tbo=
dy></table>
</body></html>

----_SiB-39bb7901476226a7-Part_1--
"""

# Extract the IP from the header
ip_address = extract_ip_from_header(email_header)

if ip_address:
    # Get location info based on the IP address
    location_info = get_location_by_ip(ip_address)
    print(f"Location information for IP {ip_address}:")
    print(location_info)
```
