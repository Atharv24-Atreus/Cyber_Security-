```python
import re
import requests

# Function to extract the IP address from email headers
def extract_ip_from_header(email_header):
    # Regular expression to find an IP address in the 'Received' field
    ip_pattern = re.compile(r"Received:.*\[(?P<ip>\d+\.\d+\.\d+\.\d+)\]")
    match = ip_pattern.search(email_header)

    if match:
        return match.group("ip")
    else:
        print("IP Address not found in the email header.")
        return None

# Function to get location information based on IP address
def get_location_by_ip(ip):
    # Use the ipinfo.io API to get location info
    api_url = f"https://ipinfo.io/{ip}/json"
    response = requests.get(api_url)

    if response.status_code == 200:
        location_data = response.json()
        location_info = {
            'IP': ip,
            'Location': location_data.get('city', 'Unknown'),
            'Region': location_data.get('region', 'Unknown'),
            'Country': location_data.get('country', 'Unknown'),
            'Organization': location_data.get('org', 'Unknown'),
        }
        return location_info
    else:
        return {"Error": "Unable to fetch location data"}

# Sample email header (you can extract this from the email you want to analyze)
email_header = """Delivered-To: atharvrastogi2409@gmail.com
Received: by 2002:a05:6830:368f:b0:71e:197c:96cd with SMTP id bk15csp118743otb;
        Thu, 30 Jan 2025 16:13:45 -0800 (PST)
X-Received: by 2002:a05:6102:38d4:b0:4b1:1565:f4f1 with SMTP id ada2fe7eead31-4b9a4ecdf5bmr8525347137.3.1738282425527;
        Thu, 30 Jan 2025 16:13:45 -0800 (PST)
ARC-Seal: i=1; a=rsa-sha256; t=1738282425; cv=none;
        d=google.com; s=arc-20240605;
        b=M4N5mpTC69egbu/YEW5rsbiLKm0iSj6HvD2A1rZ0kp38AwX3t1iDWIH6EfEH2XY2uh
         HCrbG3tGZaVn95yJdUL2rd3IOaVmMU5qZK/FOreGUiD+fjcdNF94aSPClNNt17IqmXcY
         BoxxdH0jM7yB+NLZXn+NYm+NiGYE6e9OQrQz7uDQqF4LSaOGgrgi1f5gE32kpig8cwCc
         jXnOPEqW9SmHw12RSCFSVu9zInxi04ZM/7e37FV4QxTwjr+HYSeFrWizBPcfqn6xb62/
         pdgPwtTSy176ydkIdyU9gRqOM3TpQYRCrVgRArrS/JPp51uxs0eJ/fXxUyu7b/MIdse3
         jHcw==
ARC-Message-Signature: i=1; a=rsa-sha256; c=relaxed/relaxed; d=google.com; s=arc-20240605;
        h=to:from:subject:message-id:list-id:feedback-id:precedence
         :list-unsubscribe:reply-to:date:mime-version:dkim-signature;
        bh=lv7Om3HJMEI2peWuuFlSr8TGWiN+7WokYetnGr6GMDg=;
        fh=EIETzPgcX+dT8nu9AQJUlOLR5nUShmn/gBMEwM0pG1I=;
        b=gegO4hJYtOcxFQRl+AIwpi8cyJShja4TMHzSyzERSsunlGB4yeMsffPJFwemsZv2s5
         1LZemqXWV3po3uH/hE2jpIwMaf31nYDCsMu6anmHWMfFlEZiJKhKQAV1Ib3aZL3gec8t
         bmCTg4T0a/DHJkoEHLOZhErg62yfn84+h0hglnM/BIIYOyZnjRQUkRkfK3k/sbobUiFV
         z3n76K334bpfM6P1DyQ5jtbz9kH4k4hHwzczVNUy96ugeOl3lLgzkCXc2vTEx5RQsqhI
         9V3sKMFJqv18b8/c7wEMrRs2ZhFqR99UR0IKFk98D+KZBOnvcgGJjZg16LdPd4i23X7u
         sK7g==;
        dara=google.com
ARC-Authentication-Results: i=1; mx.google.com;
       dkim=pass header.i=@kaggle.com header.s=20230601 header.b=ZbRBc3Ah;
       spf=pass (google.com: domain of 38e-bzwckeyqvwzmxt6siootm.kwui1piz3zi01woqac8houiqt.kwu@scoutcamp.bounces.google.com designates 209.85.220.69 as permitted sender) smtp.mailfrom=38e-bZwcKEYQvwzmxt6siootm.kwui1piz3zi01woqAC8Houiqt.kwu@scoutcamp.bounces.google.com;
       dmarc=pass (p=NONE sp=NONE dis=NONE) header.from=kaggle.com;
       dara=pass header.i=@gmail.com
Return-Path: <38e-bZwcKEYQvwzmxt6siootm.kwui1piz3zi01woqAC8Houiqt.kwu@scoutcamp.bounces.google.com>
Received: from mail-sor-f69.google.com (mail-sor-f69.google.com. [209.85.220.69])
        by mx.google.com with SMTPS id a1e0cc1a2514c-8669413d9c4sor838126241.3.2025.01.30.16.13.45
        for <atharvrastogi2409@gmail.com>
        (Google Transport Security);
        Thu, 30 Jan 2025 16:13:45 -0800 (PST)
Received-SPF: pass (google.com: domain of 38e-bzwckeyqvwzmxt6siootm.kwui1piz3zi01woqac8houiqt.kwu@scoutcamp.bounces.google.com designates 209.85.220.69 as permitted sender) client-ip=209.85.220.69;
Authentication-Results: mx.google.com;
       dkim=pass header.i=@kaggle.com header.s=20230601 header.b=ZbRBc3Ah;
       spf=pass (google.com: domain of 38e-bzwckeyqvwzmxt6siootm.kwui1piz3zi01woqac8houiqt.kwu@scoutcamp.bounces.google.com designates 209.85.220.69 as permitted sender) smtp.mailfrom=38e-bZwcKEYQvwzmxt6siootm.kwui1piz3zi01woqAC8Houiqt.kwu@scoutcamp.bounces.google.com;
       dmarc=pass (p=NONE sp=NONE dis=NONE) header.from=kaggle.com;
       dara=pass header.i=@gmail.com
DKIM-Signature: v=1; a=rsa-sha256; c=relaxed/relaxed;
        d=kaggle.com; s=20230601; t=1738282425; x=1738887225; dara=google.com;
        h=to:from:subject:message-id:list-id:feedback-id:precedence
         :list-unsubscribe:reply-to:date:mime-version:from:to:cc:subject:date
         :message-id:reply-to;
        bh=lv7Om3HJMEI2peWuuFlSr8TGWiN+7WokYetnGr6GMDg=;
        b=ZbRBc3AhQLIS+3VP0cZLfm0XtrzoNkoA7YAVBN6j9Ee33GvOAQEYlEol1jfhMOOYC2
         1XW352EQO8G2f3vfLRRL0RH/aYpyRmQ2WPGvPum/pS7jvtEp/e9SG7WFxRlIBIck8DSC
         mJ44z7VrtYchkfw8xWgFrMPX5O8NC4oi3GR7+oAQcU0sW9EAfaCRFv0Xmi2iUGC80pgG
         MvKEG81gIDBXAK8vyhDXdovHLxCpF9hoZJSKhBoGkhtxRbw8Ixj8zZFySrPxVjuvB4N/
         HXGwHXbYg0pOeMYsF87kZvi+bAICKNtDCVRPY9IFAbrlfhUguJOprkWi9PFQxf3diceN
         bG5A==
X-Google-DKIM-Signature: v=1; a=rsa-sha256; c=relaxed/relaxed;
        d=1e100.net; s=20230601; t=1738282425; x=1738887225;
        h=to:from:subject:message-id:list-id:feedback-id:precedence
         :list-unsubscribe:reply-to:date:mime-version:x-gm-message-state:from
         :to:cc:subject:date:message-id:reply-to;
        bh=lv7Om3HJMEI2peWuuFlSr8TGWiN+7WokYetnGr6GMDg=;
        b=mFSa5XH2YijaZ0gQlUGuN/2xwi0AYuFP+2Yl020WWEuTu8WcOVSP0e1TCGb+MzXGvH
         KTCOJCpVobBhQSoL7erEvdiTpW3eJfpQf3btalJbXgQrFDkzJY4Yc4DG76WPxWDtSUze
         yLQ5Na6G52+vNsFJ87/2E/2/02m3ZXQJCeDIfPF4f0VGGhh/jhlkStmrBNT8W5tVPdo7
         ZefRR4y7LIC608s0Nsk234Ow9UzapYKvvpj+AXvQpzmS5izZ2SsozZuih5EIc+RtzYkd
         8b8WJI6E1VHg7kXSEOhbv5ek2sn4v00eLWG9DzMCTpfWMPffzDEA3OcDUIF1e+E0MqlY
         zECQ==
X-Gm-Message-State: AOJu0Yxl3Ps8+JlRaY+HdsFjCAPSUr98G1ayJ7T0xEqLFRwsGPxVJXSf 7zSB3yuVoAngp1nL434MHSPrTZx7cKsSe4OMCvVrgVMVwSg=
X-Google-Smtp-Source: AGHT+IGx0DmCX5aGhL8gTDuQC+NWYVbMxHzy4GhjvxlOZCDlR78rljLTsy30ilvGoW6QMm1Mh8352r8=
MIME-Version: 1.0
X-Received: by 2002:a05:6e02:1caa:b0:3cf:f88b:b51a with SMTP id e9e14a558f8ab-3cffe3a6ed0mr91018095ab.2.1738272753703; Thu, 30 Jan 2025 13:32:33 -0800 (PST)
Date: Thu, 30 Jan 2025 13:32:33 -0800
Reply-To: Kaggle <noreply@kaggle.com>
X-Google-Id: 4343752
List-Unsubscribe: <https://notifications.google.com/g/p/ANiao5rwmqG6f4OLFhK6VpHCWyZX3JYxo76mJPLpoXc-v19zR4eVo6ogHpgHNwl-Qip3G7bIsD5t7rWpsw_m_XbGYEK4PFRp5AOMmRkb_JOH3cQ96kJDWNb6sqIh6Xmvynh-g6RrIu-HJbJHtyg_BhgoHrtBXofV9wBYv-8j0jcVLuXBoVvPLP3Ad007bVxVmrd_00trosLmVFSthcgEcOWci1e4HQ>
Precedence: bulk
Feedback-ID: P2603-0:C20336619:M110876066-en-US:gamma
List-Id: <d3d572f2a797bc005fbf0c631ac17de03856ef11.google.com>
X-Notifications: GAMMA:<e132d201ca512fc6db56b8d8d50c17f4889df0a1-20336619-111571626@google.com>
X-Notifications-Bounce-Info: Ab52iKkihgsT-xSs9L6e-97gF-7SiNZ5NGIc6bU9A2UqhbujX-lnf1UP-_Mg_TQMeBhTGNwy9SgLEa5PTANCnqLaE_7rDG7lzj4xqI-rIQLp5URi1rbb2uGokeUnpQU28B1vORI4schN6CLmL131Zkf5Z6pJaH7E3C4C18sPGpcs41A5GIbNFr0XUtBbU2h6ixspKW-bYDDOjsGhIFxHdJTnVHCEghB-itZqhn-gIrYhwqOp1DIcb01TSHOfoGtR0jVi_cj_aye5ZamQlp0uIsUgClfFD92zDFnthXgEGgYDgoxXENJeCzdKhfuBr0jF1aAOBLhg3pXwI0rNWnY-qtUPZWpANlGfUCWj0aJaAL6BgPtFj28-R6tLvNYU7ttsVd84Wsk8oLOjzQtg3vJxyptva_tJOaSsgyisix2AAOcNXVkHQ35VXHj5yqYLQbMO8Z1c-9exq8AICIdx4O4R9Ofb9s5py74qsMXzqGrYA1etKsxxKQQnhasm8qv9JK3DESzmgDo6zRHq4rgY0j4bWI8jKvDuxEQZgyhGNjAwNjA0MDQxNTM1NTk2OTMzMg
Message-ID: <e132d201ca512fc6db56b8d8d50c17f4889df0a1-20336619-111571626@google.com>
Subject: [Model Launch] Mistral Small 3 Is Now on Kaggle!
From: Kaggle <noreply@kaggle.com>
To: atharvrastogi2409@gmail.com
Content-Type: multipart/alternative; boundary="00000000000001dc5d062cf32acd"

--00000000000001dc5d062cf32acd
Content-Type: text/plain; charset="UTF-8"; format=flowed; delsp=yes

Kaggle








Hi Atharv Rastogi 09,

Welcome Mistral Small 3 to Kaggle!

Mistral Small 3 is a latency-optimized 24B-parameter model, designed to
deliver strong performance with high efficiency. Licensed under Apache 2.0,
it provides a flexible and scalable solution for a wide range of AI
applications. It achieves over 81% accuracy on MMLU and delivers
low-latency inference at 150 tokens per second, making it well-suited for
real-time AI tasks.





Access Mistral Small 3



This model is built for conversational AI, function calling, and
fine-tuning for specialized tasks. It enables fast-response virtual
assistants, agentic workflows, and local deployment on consumer hardware.
With both pretrained and instruction-tuned checkpoints, Mistral Small 3 is
a strong foundation for AI development.

We're excited to see what you build with it!

Brenda Flynn

Kaggle Partnerships Lead



Kaggle, Inc 1600 Amphitheatre Pkwy Mountain View, CA 94043

This email was sent to atharvrastogi2409@gmail.com because you indicated
that you'd like to receive news and updates about Kaggle. If you don't want
to receive these emails in the future, please unsubscribe here. You can
also change your preferences on your account's profile page by logging in
at kaggle.com.





--00000000000001dc5d062cf32acd
Content-Type: text/html; charset="UTF-8"
Content-Transfer-Encoding: quoted-printable

<!doctype html>
<html xmlns=3Dhttp://www.w3.org/1999/xhtml xmlns:v=3Durn:schemas-microsoft-=
com:vml xmlns:o=3Durn:schemas-microsoft-com:office:office lang=3Den xml:lan=
g=3Den>
  <head>
    <meta http-equiv=3DContent-Type content=3D"text/html; charset=3DUTF-8">
    <title> [Model Launch] Mistral Small 3 Is Now on Kaggle!
</title>
    <style type=3Dtext/css>
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 500;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa2JL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* cyrillic */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 500;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa0ZL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* greek-ext */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 500;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa2ZL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* greek */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 500;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa1pL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* vietnamese */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 500;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa2pL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* latin-ext */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 500;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa25L7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* latin */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 500;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa1ZL7W0Q5nw.woff2)
          format("woff2");
      }
      /* cyrillic-ext */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 600;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa2JL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* cyrillic */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 600;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa0ZL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* greek-ext */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 600;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa2ZL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* greek */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 600;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa1pL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* vietnamese */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 600;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa2pL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* latin-ext */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 600;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa25L7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* latin */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 600;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa1ZL7W0Q5nw.woff2)
          format("woff2");
      }
      /* cyrillic-ext */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 700;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa2JL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* cyrillic */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 700;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa0ZL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* greek-ext */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 700;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa2ZL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* greek */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 700;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa1pL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* vietnamese */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 700;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa2pL7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* latin-ext */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 700;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa25L7W0Q5n-wU.woff2)
          format("woff2");
      }
      /* latin */
      @font-face {
        font-family: "Inter";
        font-style: normal;
        font-weight: 700;
        font-display: swap;
        src: url(https://fonts.gstatic.com/s/inter/v2/UcC73FwrK3iLTeHuS_fvQ=
tMwCp50KnMa1ZL7W0Q5nw.woff2)
          format("woff2");
      }
      html {
        background: #f5f5f5;
        font-family: "Inter", sans-serif;
      }

      /* WINDOWS 10 MAIL LINK FIX */
      a {
        text-decoration: none;
      }

      /* iOS GMAIL WEB FONT SIZE FIX */
      .gmail-fix {
        display: none;
        display: none !important;
      }

      /* iOS GMAIL APP WIDTH FIX */
      u + #new-gmail-fix {
        display: block !important;
      }

      /* /////////////////// FIX */
      .cta-hover:hover {
        background-color: #639420 !important;
        color: #ffffff !important;
      }

      /* /////////////////// FIX */
      /* CTA ROLLOVER IF SUPPORTED */
      .cta-hover:hover {
        color: #ffffff !important;
        background-color: #639420 !important;
      }

      .overline {
        font-family: "Inter", sans-serif;
        font-size: 10px;
        padding: 0;
        margin: 0;
        color: #202124;
        letter-spacing: 1px;
        text-transform: capitalize;
      }

      h1 {
        font-family: "Inter", sans-serif;
        font-weight: 700;
        font-size: 36px;
        color: #202124;
        letter-spacing: -0.02px;
        margin: 0;
      }

      .body1 {
        font-family: Inter-Medium;
        font-size: 16px;
        color: #3c4043;
        letter-spacing: -0.02px;
        line-height: 24px;
      }

      h2 {
        font-family: "Inter", sans-serif;
        font-weight: 700;
        font-size: 28px;
        color: #202124;
        letter-spacing: -0.02px;
        margin: 38px 0 0 0;
      }

      body {
      }
      .side-padding {
          padding: 0 92px 0 92px;
          margin: 0;
        }

      @media screen and (min-width: 600px) {
        .side-padding {
          padding: 0 92px 0 92px;
          margin: 0;
        }
      }

      @media screen and (max-width: 600px) {
        .side-padding {
          padding: 0 24px 0 24px;
          margin: 0;
        }
      }
    </style>

    <!--[if mso]>
      <style type=3D"text/css">
        table {
          border-collapse: collapse;
          mso-table-lspace: 0pt;
          mso-table-rspace: 0pt;
        }
        table td {
          border-collapse: collapse;
        }
        .mso-cta {
          padding: 10px 0 10px 0 !important;
          font-weight: bold !important;
        }
      </style>
    <![endif]-->
  </head>

  <body bgcolor=3D#F5F5F5 style=3D"padding: 0; margin: 0; background: #f5f5=
f5">
    <center>
      <!-- OUTER WRAPPER -->
      <table align=3Dcenter bgcolor=3D#F5F5F5 border=3D0 cellpadding=3D0 ce=
llspacing=3D0 width=3D100% role=3Dpresentation>
        <tbody>
          <tr>
            <td align=3Dcenter style=3D"padding: 0px 0 0 0">
              <!-- CONTENT CONTAINER -->
              <table align=3Dcenter border=3D0 cellpadding=3D0 cellspacing=
=3D0 width=3D600 role=3Dpresentation>
                <tbody>
                  <tr bgcolor=3D#F5F5F5 height=3D32>
                    <td style=3D"padding: 0 0 0 0; height: 32px"></td>
                  </tr>
                  <tr bgcolor=3D#F5F5F5>
                    <td align=3Dcenter style=3D"padding: 0 0 0 0" width=3D6=
00px>
                      <img alt=3DKaggle border=3D0 src=3Dhttps://storage.go=
ogleapis.com/kaggle-media/Images/Kaggle_Blue.png alt=3D"=E2=80=9CKaggle=E2=
=80=9D" style=3D"
                          display: block;
                          outline: none;
                          height: auto;
                          height: 40px;
                          width: 104px;
                        " width=3D104>
                    </td>
                  </tr>
                  <tr bgcolor=3D#F5F5F5 height=3D32>
                    <td style=3D"padding: 0 0 0 0; height: 32px"></td>
                  </tr>
                  <!-- TEXT CONTENT CONTAINER -->
                  <tr bgcolor=3D#FFFFFF>
                    <td align=3Dcenter style=3D"padding: 0 0 0 0" width=3D6=
00px>
                      <table align=3Dcenter border=3D0 cellpadding=3D0 cell=
spacing=3D0 width=3D600 role=3Dpresentation>
                        <tbody>
                          <tr bgcolor=3D#FFFFFF height=3D36>
                            <td style=3D"padding: 0 0 0 0; height: 36px"></=
td>
                          </tr>
                          <tr bgcolor=3D#FFFFFF>
                            <td class=3Dside-padding style=3D"padding: 0 44=
px 0 44px">
                             =20
                       =20

                              <p style=3D"
                                  font-family: &#39;Inter&#39;, sans-serif;
                                  font-weight: 500;
                                  font-size: 16px;
                                  color: #3c4043;
                                  letter-spacing: -0.02px;
                                  line-height: 24px;
                                ">
                                Hi Atharv Rastogi 09,
                              </p>
                           <p style=3D"
                                  font-family: &#39;Inter&#39;, sans-serif;
                                  font-weight: 500;
                                  font-size: 16px;
                                  color: #3c4043;
                                  letter-spacing: -0.02px;
                                  line-height: 24px;
                                ">Welcome Mistral Small 3 to Kaggle!
<br>
<br>
Mistral Small 3 is a latency-optimized 24B-parameter model, designed to del=
iver strong performance with high efficiency. Licensed under Apache 2.0, it=
 provides a flexible and scalable solution for a wide range of AI applicati=
ons. It achieves over 81% accuracy on MMLU and delivers low-latency inferen=
ce at 150 tokens per second, making it well-suited for real-time AI tasks.=
=20


</p>
      =20
                 =20
   =20
                           =20
                            </td>
                          </tr>
                        </tbody>
                      </table>
                    </td>
                  </tr>



             =20
                  <tr bgcolor=3D#FFFFFF>
                    <td align=3Dcenter>
                      <div>
                        <!--[if mso]>
                          <v:roundrect
                            xmlns:v=3D"urn:schemas-microsoft-com:vml"
                            xmlns:w=3D"urn:schemas-microsoft-com:office:wor=
d"
                            href=3D"https://www.kaggle.com/models/mistral-a=
i/mistral-small-24b?utm_medium=3Demail&utm_source=3Dgamma&utm_campaign=3Dmo=
delslaunch-mistral-small3"=20
                            style=3D"
                              height: 36px;
                              v-text-anchor: middle;
                              width: 200px;
                            "
                            arcsize=3D"50%"
                            strokecolor=3D"#e6e6e8"
                            fillcolor=3D"#fafafb"
                          >
                            <w:anchorlock />
                            <center
                              style=3D"
                                color: #2f353e;
                                font-family: sans-serif;
                                font-size: 13px;
                                font-weight: bold;
                              "
                            >
Access Mistral Small 3                            </center>
                          </v:roundrect> <!
                        [endif]--><a href=3Dhttps://notifications.google.co=
m/g/p/ANiao5oYa_YpcFkxL_DUsPBu5Zi6D7bF_BfTESXxUoXXJ9qWYEDEzC81Ba3UGsvZfP75s=
8M5H98mIrRxK_KE0u9WTJ-MwdihwsrP08Ds1DjOltCR1qbt14qTQlKk0SpSzfzJPTHIMhxr03pY=
abldgw-U3wIQFil2ei4pQqa7bZudYfbpkI248Ff6g2u2wNdnWKmNdNXj9dc1wmCAt4Gdu_pC5NI=
zpYsnwHA1pqhevjFDYiV0rMPNOnG6oKHOAKRXagMU83WxMw style=3D"
                            background-color: #000000;
                            padding: 11px 24px 11px 24px;
                            margin: 10px 0 0px 0;
                            border-radius: 20px;
                            font-family: &#39;Inter&#39;, sans-serif;
                            font-weight: 600;
                            font-size: 14px;
                            color: #ffffff;
                            letter-spacing: 0;
                            display: inline-block;
                            text-align: center;
                            text-decoration: none;
                            -webkit-text-size-adjust: none;
                            mso-hide: all;
                          " title=3D"Access Mistral Small 3">     Access Mi=
stral Small 3=20
</a>
                      </div>
                    </td>
                  </tr>



            =20
                  <tr bgcolor=3D#FFFFFF>
                    <td class=3Dside-padding style=3D"padding: 0 44px 0 44p=
x">
 <p style=3D"
                          font-family: &#39;Inter&#39;, sans-serif;
                          font-weight: 500;
                          font-size: 16px;
                          padding-bottom: 40px;
                          color: #3c4043;
                          letter-spacing: -0.02px;
                          line-height: 24px;
                        ">
This model is built for conversational AI, function calling, and fine-tunin=
g for specialized tasks. It enables fast-response virtual assistants, agent=
ic workflows, and local deployment on consumer hardware. With both pretrain=
ed and instruction-tuned checkpoints, Mistral Small 3 is a strong foundatio=
n for AI development.

<br>
<br>

We=E2=80=99re excited to see what you build with it!

<br>
<br>
Brenda Flynn <br><br>Kaggle Partnerships Lead

                      </p>
                    </td>
                  </tr>


                  <tr>
                    <td class=3Dside-padding style=3D"padding: 0 24px 0 24p=
x">
                      <p style=3D"
                          font-family: &#39;Inter&#39;, sans-serif;
                          font-weight: 500;
                          font-size: 12px;
                          color: #5f6368;
                          padding-top: 30px;
                          letter-spacing: 0;
                          text-align: center;
                          line-height: 20px;
                        ">
                        Kaggle, Inc 1600 Amphitheatre Pkwy Mountain View, C=
A
                        94043
                      </p>
                      <p style=3D"
                          font-family: &#39;Inter&#39;, sans-serif;
                          font-weight: 500;
                          font-size: 12px;
                          color: #5f6368;
                          letter-spacing: 0;
                          text-align: center;
                          line-height: 20px;
                        ">
                        This email was sent to atharvrastogi2409@gmail.com =
because you
                        indicated that you&#39;d like to receive news and u=
pdates
                        about Kaggle. If you don&#39;t want to receive thes=
e emails
                        in the future, please
                        <a href=3Dhttps://notifications.google.com/g/p/ANia=
o5rwmqG6f4OLFhK6VpHCWyZX3JYxo76mJPLpoXc-v19zR4eVo6ogHpgHNwl-Qip3G7bIsD5t7rW=
psw_m_XbGYEK4PFRp5AOMmRkb_JOH3cQ96kJDWNb6sqIh6Xmvynh-g6RrIu-HJbJHtyg_BhgoHr=
tBXofV9wBYv-8j0jcVLuXBoVvPLP3Ad007bVxVmrd_00trosLmVFSthcgEcOWci1e4HQ style=
=3D"color: #000000; text-decoration: underline" title=3D"Unsubscribe here">=
unsubscribe here</a>. You can also change your preferences on your
                        account&#39;s profile page by logging in at
                        <a href=3Dhttps://notifications.google.com/g/p/ANia=
o5rsAP2sACbmpjurJKXU0ck9DH-WKNbHUAvAftdBj9pNdYlysGYXwaQDudru--rl8isj5WI2q7r=
0yA7AZUfPQehJgctdPXCN462oiUFsDPAf_9zt style=3D"color: #000000; text-decorat=
ion: underline" title=3DKaggle.com>kaggle.com.</a>
                      </p>
                    </td>
                  </tr>
                </tbody>
              </table>
            </td>
          </tr>
        </tbody>
      </table>
    </center>
  <img alt=3D"" height=3D1 width=3D3 src=3Dhttps://notifications.google.com=
/g/img/ANiao5oaPadP7_bQnNESel9mRHybyfX4g24eHrsQNAGtKO4JTuAM5riA5hvb1g94oeIk=
5g7rdMW5RzqTAaN3Yx85BQ4UPiNayue0ByviNJPwuld3OWcsW9KRx9wTX-6r_c71K2vY6A.gif>=
</body>
</html>
--00000000000001dc5d062cf32acd--
"""

# Extract the IP from the header
ip_address = extract_ip_from_header(email_header)

if ip_address:
    # Get location info based on the IP address
    location_info = get_location_by_ip(ip_address)
    print(f"Location information for IP {ip_address}:")
    print(location_info)
```

```python
from scapy.all import rdpcap, TCP, Raw, IP  # Import required Scapy modules
import re
from collections import defaultdict

# Path to the PCAP file
pcap_file_path = "fake_email_traffic.pcap"

# Read packets from the PCAP file
packets = rdpcap(pcap_file_path)

# SMTP Command Patterns
smtp_regex = {
    "EHLO": re.compile(rb"EHLO"),
    "MAIL FROM": re.compile(rb"MAIL FROM:\s*<([^>]+)>"),
    "RCPT TO": re.compile(rb"RCPT TO:\s*<([^>]+)>"),
    "DATA": re.compile(rb"DATA"),
    "DOT": re.compile(rb"^\.\r\n"),  # End of DATA section
    "QUIT": re.compile(rb"QUIT")
}

# Storage for SMTP sessions
sessions = defaultdict(list)  # {source_ip: [commands]}
malformed_sessions = []  # List of problematic sessions

# Process packets to extract SMTP commands
for packet in packets:
    if packet.haslayer(IP) and packet.haslayer(TCP) and packet.haslayer(Raw):
        payload = packet[Raw].load
        src_ip = packet[IP].src
        dst_ip = packet[IP].dst

        for command, pattern in smtp_regex.items():
            if pattern.search(payload):  # Check if command is in payload
                sessions[src_ip].append(command)  # Store command in session

# Analyze SMTP sessions for irregularities
for src_ip, commands in sessions.items():
    if "MAIL FROM" in commands and "RCPT TO" not in commands:
        malformed_sessions.append((src_ip, "Missing RCPT TO after MAIL FROM"))

    if "DATA" in commands and "DOT" not in commands:
        malformed_sessions.append((src_ip, "DATA command without termination (missing '.')"))

    if "EHLO" in commands and "QUIT" not in commands:
        malformed_sessions.append((src_ip, "Session missing QUIT command"))

    if any(cmd not in ["EHLO", "MAIL FROM", "RCPT TO", "DATA", "DOT", "QUIT"] for cmd in commands):
        malformed_sessions.append((src_ip, "Unexpected SMTP command sequence"))

# Print the irregular SMTP sessions
print("\n🚨 **Malformed or Incomplete SMTP Sessions Detected:**")
if malformed_sessions:
    for src_ip, issue in malformed_sessions:
        print(f"⚠️ {src_ip}: {issue}")
else:
    print("✅ No issues detected")
```
